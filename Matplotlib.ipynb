```python
bike_sales = sales[sales['product_subcategory'].isin(['Road Bikes', 'Mountain Bikes', 'Touring Bikes'])]
```

```python
bike_sales_grp = bike_sales.groupby('product_subcategory')['SalesAmount'].sum().reset_index()
```

```python
accessory_sales = sales[~sales['product_subcategory'].isin(['Road Bikes', 'Mountain Bikes', 'Touring Bikes'])]
```

```python
accessory_sales_grp = accessory_sales.groupby('product_subcategory')\
                    ['SalesAmount'].sum().reset_index().sort_values(by = 'SalesAmount')
```

```python
accessory_sales_grp
```

```python
plt.pie(bike_sales_grp['SalesAmount'], labels = bike_sales_grp['product_subcategory'])
```

```python
plt.barh(accessory_sales_grp['product_subcategory'], accessory_sales_grp['SalesAmount'])
```

```python
sales['OrderDate'] = pd.to_datetime(sales['OrderDateKey'], format = '%Y%m%d')
sales['MonthYear'] = pd.to_datetime(sales['OrderDate'].dt.strftime('%Y-%m'))
sales
```

```python
monthly_sales = sales.groupby('MonthYear')['SalesAmount'].sum().reset_index()
daily_sales = sales.groupby('OrderDate')['SalesAmount'].sum().reset_index()
```

```python
plt.plot(monthly_sales['MonthYear'], monthly_sales['SalesAmount'])
plt.show()
```

```python
plt.plot(daily_sales['OrderDate'], daily_sales['SalesAmount'])
plt.show()
```

```python
country_sales = sales.groupby('SalesTerritoryCountry')['SalesAmount']\
                .sum().reset_index().sort_values(by='SalesAmount',ascending=False)
```

```python
plt.bar(country_sales['SalesTerritoryCountry'], country_sales['SalesAmount'])
```

```python
import matplotlib.pyplot as plt

# Create a 2x2 grid of plots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
plt.tight_layout(pad=4)

# Plot 1: Scatter plot
axs[0, 0].scatter(world['Healthy life expectancy'], world['Score'], alpha=0.5, color='green')
axs[0, 0].grid(True)
axs[0, 0].set_xlabel('Healthy life expectancy')
axs[0, 0].set_ylabel('Happiness Score')
axs[0, 0].set_title('Life Expectancy vs. Happiness')

# Plot 2: Bar plot
countries = ['USA', 'Japan', 'UK', 'Germany']
profits = [600000, 300000, 200000, 200000]
axs[0, 1].bar(countries, profits, color=['#003f5c', '#2f4b7c', '#665191', '#a05195'])
axs[0, 1].set_ylabel('Profits')
axs[0, 1].set_title('Profits by Country')

# Plot 3: Pie chart
axs[1, 0].pie(profits, labels=countries, autopct='%1.1f%%', explode=[0, 0, 0, 0.1])
axs[1, 0].set_title('Profit Distribution')

# Plot 4: Line plot
axs[1, 1].plot(eth.index, eth['High'], color = 'orange')
axs[1, 1].set_xlabel('Date')
axs[1, 1].set_ylabel('High')
axs[1, 1].set_title('Ethereum Price Chart')
axs[1, 1].tick_params(axis='x', rotation=45)

# Adjust layout and show the plots
plt.show()
```
